```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("download2/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_5728786.csv", skiprows=4)
data
```

```python
# data = data.drop(["Indicator Name", "Indicator Code"], axis=1)

keep = ["European Union", "Germany", "Euro area", "Europe & Central Asia"]
to_drop = []

for indx, row in data.iterrows():
    # print(row["Country Name"])
    if row["Country Name"] not in keep:
        to_drop.append(indx)

data.drop(index=to_drop, inplace=True)
data
```

```python
for indx, row in data.iterrows():
    name = row["Country Name"]
    data_row = row.drop(["Country Name", "Country Code", "Unnamed: 67"])
    plt.plot(data_row, label=name)
plt.legend()
plt.show()
```
