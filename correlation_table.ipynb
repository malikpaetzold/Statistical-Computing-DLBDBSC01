```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

sns.set_style("white")
```

```python
data = pd.DataFrame({"Einschulungsrate Sekundär": 0.729, "Einschulungsrate Tetiär": 0.747, "Geburtenrate": -0.708, "Lebenserwartung": 0.787, "DPT-Immunisierung": 0.598, "Elektrizität": 0.583}, index=["BIP pro Kopf"])
print(data)

cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(data, annot=True, linewidths=.5, ax=ax, vmax=1, vmin=-1, cmap=cmap, cbar=True, square=True, cbar_kws={"orientation": "horizontal", "location": "top"})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
# plt.title("Ergebnisse des Korrelationskoeffizienten r")
plt.tight_layout()
plt.savefig("correlation_per_class.png", dpi=300)
```
